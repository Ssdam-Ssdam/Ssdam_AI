```python
# 1000개의 데이터가 모이도록
import os
import urllib.request
import json

item = "김치냉장고"
client_id = "20LpqdMBabcnhllH004Y"
client_secret = "LK4enA8gPh"
encText = urllib.parse.quote(item)

# 이미지 저장 경로
savePath = "C:/Users/lej55/OneDrive/바탕 화면/data/장롱"
os.makedirs(savePath, exist_ok=True)  # 디렉토리가 없으면 생성

# 1000개 이미지 다운로드
total_images = 0
max_images = 1000
start = 1  # 검색 결과의 시작점

while total_images < max_images:
    url = f"https://openapi.naver.com/v1/search/image?display=100&start={start}&query={encText}"

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)
    
    try:
        response = urllib.request.urlopen(request)
        rescode = response.getcode()

        if rescode == 200:
            response_body = response.read()
            result = json.loads(response_body)
            img_list = result['items']

            for i, img_data in enumerate(img_list, start=total_images + 1):
                # 이미지 링크 확인
                img_url = img_data['link']
                print(f"Downloading image {i}: {img_url}")

                # 저장 파일명 및 경로
                file_name = os.path.join(savePath, item + str(i) + '.jpg')

                # 이미지 URL이 유효한 형식인지 확인
                if img_url.endswith(('.jpg', '.png', '.jpeg')):
                    try:
                        # 이미지 다운로드 요청
                        urllib.request.urlretrieve(img_url, file_name)
                        print(f"Downloaded: {file_name}")
                        total_images += 1
                        if total_images >= max_images:
                            break
                    except Exception as e:
                        print(f"Failed to download {img_url}: {e}")
                else:
                    print(f"Invalid image URL: {img_url}")
            
            # 페이지 번호 업데이트
            start += 100  # 페이지 번호 증가

        else:
            print("Error Code:" + rescode)
            break

    except Exception as e:
        print(f"Error occurred while fetching data: {e}")
        break

# 다운로드 완료 시 출력
print(f"-------- Download completed: {total_images} images --------")
```

```python
import os
import urllib.request
import json

# 네이버 API 인증 정보
client_id = "20LpqdMBabcnhllH004Y"
client_secret = "LK4enA8gPh"

# 레이블 리스트
labels = [
    "장롱", "서랍장", "쇼파", "식탁", "책상", "책장", "책꽂이", "침대", "의자", "화장대", 
    "TV 받침", "진열대", "냉장고", "김치냉장고", "텔레비전", "세탁기", "에어컨", 
    "선풍기", "피아노", "거울", "자전거", "빨래건조대", "안마의자"
]

# 저장 경로 설정
base_path = "C:/Users/lej55/OneDrive/바탕 화면/p_ssdam/data"

# 다운로드 함수
def download_images(label):
    label_path = os.path.join(base_path, label)
    os.makedirs(label_path, exist_ok=True)

    total_images = 0
    start = 1

    while total_images < 1000:
        # API 요청 URL 생성
        encText = urllib.parse.quote(label)
        url = f"https://openapi.naver.com/v1/search/image?display=100&start={start}&query={encText}"

        # API 요청
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()

        if rescode == 200:
            response_body = response.read()
            result = json.loads(response_body)
            img_list = result.get('items', [])

            if not img_list:  # 더 이상 결과가 없을 경우 중단
                print(f"No more results for {label}.")
                break

            for img in img_list:
                if total_images >= 1000:
                    break

                img_url = img['link']
                if img_url.endswith(('.jpg', '.png', '.jpeg')):
                    try:
                        file_name = os.path.join(label_path, f"{label}_{total_images + 1}.jpg")
                        urllib.request.urlretrieve(img_url, file_name)
                        total_images += 1
                        print(f"Downloaded: {file_name}")
                    except Exception as e:
                        print(f"Failed to download {img_url}: {e}")
                else:
                    print(f"Invalid image URL: {img_url}")
        else:
            print(f"Error Code: {rescode} for label {label}")
            break

        start += 100  # 다음 페이지 시작점

    print(f"Finished downloading for {label}. Total images: {total_images}")

# 모든 레이블에 대해 실행
for label in labels:
    print(f"Starting download for {label}...")
    download_images(label)

print("-------- All downloads completed --------")
```
